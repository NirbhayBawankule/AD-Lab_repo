```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import seaborn as sns

%matplotlib inline
```

```python
data = fetch_california_housing()
df = pd.DataFrame(data.data, columns=data.feature_names)
df["TargetValue"] = data.target  # median house value in 100k USD

df.head()
```

```python
X = df[["MedInc"]]      # feature
y = df["TargetValue"]   # target
```

```python
model = LinearRegression()
model.fit(X, y)

pred = model.predict(X)
```

```python
mse = mean_squared_error(y, pred)
r2 = r2_score(y, pred)

print("Mean Squared Error:", mse)
print("R² Score:", r2)
```

```python
plt.figure(figsize=(7,5))

sns.scatterplot(x = y ,y= pred, color='pink')
plt.axline((0, 0), slope=1, color='green', linestyle='--', linewidth=2)
plt.xlabel("Median Income")
plt.ylabel("Median House Value (100k USD)")
plt.title("Linear Regression on California Housing Dataset")
plt.legend()

# Optional: your name in top-right
plt.text(0.95, 0.95, "Nirbhay Bawankule\nRoll No: 23053222",
         fontsize=10, color='black',
         ha='right', va='top',
         transform=plt.gca().transAxes)

plt.show()
```
